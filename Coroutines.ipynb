# Generator function

```python
def say(start):
    msg = start
    for _ in range(10):
        word = yield msg
        msg += ' ' + word

speak = say('Hello')
speak.send(None) # prime generator
print(speak.send('you'))
print(speak.send('there'))
print(speak.send('behind'))
print(speak.send('your'))
print(speak.send('computer'))
print(speak.send('screen'))
print(speak.send('staring'))
print(speak.send('at'))
print(speak.send('this.'))
# print(speak.send('END'))
print(speak.close())
```
