# Data Classes

## collections.namedtuple

```python
from collections import namedtuple

Coordinate = namedtuple('Coordinate', 'lat lon')

moscow = Coordinate(55.756, 37.617)

inspect(moscow, dunder=True)
moscow == Coordinate(lat=55.756, lon=37.617)  # __eq__ compares all attributes
```

## typing.NamedTuple

```python
import typing

# Coordinate = typing.NamedTuple('Coordinate', [('lat', float), ('lon', float)])
# or shorter:
Coordinate = typing.NamedTuple('Coordinate', lat=float, lon=float)

moscow = Coordinate(55.756, 37.617 )# immutable instance

inspect(moscow, dunder=True)    
print(typing.get_type_hints(Coordinate))
```

```python
from typing import NamedTuple

class Coordinate(NamedTuple):
    lat: float
    lon: float

moscow = Coordinate(55.756, 37.617) # immutable instance

inspect(moscow, dunder=True)
```

## @dataclasses.dataclass

```python
from dataclasses import dataclass
from rich.traceback import install
install(show_locals=True)
from rich import console

@dataclass(frozen=True)
class Coordinate:
    lat: float
    lon: float

moscow = Coordinate(55.756, 37.617) # mutable instance, protected with 'frozen=True'

 
inspect(moscow, dunder=True)
```
